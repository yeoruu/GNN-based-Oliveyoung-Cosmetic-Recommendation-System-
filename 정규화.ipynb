```python
# reviews와 ingredients 데이터셋에서 같은 product_id의 product_name 일치 여부 확인

import pandas as pd

# 데이터 로드
reviews = pd.read_csv("crawling_data/review_only_100_data/table1_essence_reviews_259.csv")
ingredients = pd.read_csv("crawling_data/review_only_100_data/table3_essence_ingredients_259.csv")

print(f"reviews 리뷰 수: {len(reviews)}")
print(f"reviews 고유 product_id 수: {reviews['product_id'].nunique()}")
print(f"ingredients 상품 수: {len(ingredients)}")

# 각 데이터셋에서 product_id별 product_name 추출
reviews_names = reviews[['product_id', 'product_name']].drop_duplicates()
ingredients_names = ingredients[['product_id', 'product_name']].drop_duplicates()

print(f"\nreviews 고유 (product_id, product_name) 조합: {len(reviews_names)}")
print(f"ingredients 고유 (product_id, product_name) 조합: {len(ingredients_names)}")

# 두 데이터셋 병합하여 비교
merged = reviews_names.merge(
    ingredients_names, 
    on='product_id', 
    how='outer',
    suffixes=('_reviews', '_ingredients')
)

# product_name이 다른 경우 찾기
mismatched = merged[merged['product_name_reviews'] != merged['product_name_ingredients']]

print(f"\n=== 검사 결과 ===")
if len(mismatched) == 0:
    print("✅ 모든 product_id의 product_name이 일치합니다!")
else:
    print(f"❌ product_name이 일치하지 않는 product_id: {len(mismatched)}개")
    print("\n일치하지 않는 항목:")
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_colwidth', None)
    print(mismatched)
    
    # 상세 정보
    print("\n=== 불일치 상세 ===")
    for idx, row in mismatched.iterrows():
        print(f"\nProduct ID: {row['product_id']}")
        print(f"  Reviews:     {row['product_name_reviews']}")
        print(f"  Ingredients: {row['product_name_ingredients']}")

# reviews에만 있는 product_id
only_in_reviews = merged[merged['product_name_ingredients'].isna()]
if len(only_in_reviews) > 0:
    print(f"\n⚠️ reviews에만 있고 ingredients에 없는 product_id: {len(only_in_reviews)}개")
    print(only_in_reviews[['product_id', 'product_name_reviews']])

# ingredients에만 있는 product_id
only_in_ingredients = merged[merged['product_name_reviews'].isna()]
if len(only_in_ingredients) > 0:
    print(f"\n⚠️ ingredients에만 있고 reviews에 없는 product_id: {len(only_in_ingredients)}개")
    print(only_in_ingredients[['product_id', 'product_name_ingredients']])
```

```python
import re
import pandas as pd

# =========================
# 1) 제품명 정규화 함수 (브랜드명 제거 포함)  -> 이번엔 brand 미사용이라 그대로 둠
# =========================
PROMO_PATTERNS = [
    r"\[.*?\]",                          # [1월올영픽/리뉴얼], [증량기획] 등
    r"\(.*?\)",                          # (…증정), (…기획) 등
    r"(?i)\bset\b",                      # set
    r"기획팩|기획세트|기획|한정기획|한정|증량기획|증량|리뉴얼|어워즈|올영픽|온라인몰\s*단독기획|단독기획|단품",
    r"더블\s*기획|트리플\s*기획|1\+1|2\+1|3\+1",
    r"증정|사은품|증정품|추가\s*증정|리필|리필팩|대용량|미니|샘플|사은|특가|한정판",
    r"\d+\s*종\s*(?:중\s*)?택\s*1",
    r"(?:\d+\s*종\s*)?중\s*택\s*1",
    r"\d+\s*입|[0-9]+\s*개입|[0-9]+\s*개", # 2입, 3개 등(팩 구성 표기)
    # 1) (단품/기획/리필/…) 택 1 / 중 택 1
    r"(?:단품|기획|세트|본품|리필|증정품|사은품)(?:\s*/\s*(?:단품|기획|세트|본품|리필|증정품|사은품))*\s*(?:중\s*)?택\s*1",
    r"\d+\s*중\s*택\s*1",     # 2중 택 1 / 2 중 택 1
    r"\b\d+\b$",              # 맨 끝에 단독 숫자(예: ... 2) 제거
    r"\d+\s*종\b",            # 2종 / 3종 같은 구성 표기 제거 (택1 없어도 제거)
    r"\b세트\b",              # 한글 '세트' 단독 제거

    # 2) 2중 택 / 3중택1 / 2 중 택 1  (숫자 + '중' + '택(1)')
    r"\d+\s*중\s*택\s*(?:1)?",

    # 3) 그냥 '택 1' / '택1' 단독으로도 남으면 제거 (옵션 꼬리 제거용)
    r"\b택\s*1\b",
    r"\+.*$",                           # "+100ml 추가증정" 같은 뒤쪽 꼬리 제거
]


VOLUME_PATTERNS = [
    r"\d+(?:\.\d+)?\s*(?:ml|mL|ML)\b",
    r"\d+(?:\.\d+)?\s*(?:g|G)\b",
    r"\d+(?:\.\d+)?\s*(?:oz|OZ)\b",
]

def normalize_product_name(raw_name: str, brand: str = None) -> str:
    """
    올리브영 상품명에서 '대표 제품명'만 남기는 정규화 함수.
    (룰은 그대로. 이번 작업에서는 brand 인자를 아예 안 넘김)
    """
    if pd.isna(raw_name):
        return ""
    s = str(raw_name).strip()

    # 1) 대괄호/괄호/프로모션 패턴 제거
    for pat in PROMO_PATTERNS:
        s = re.sub(pat, " ", s)

    # 2) 용량/중량 제거
    for pat in VOLUME_PATTERNS:
        s = re.sub(pat, " ", s)

    # 3) 흔한 구분자 정리
    s = s.replace("·", " ").replace("/", " ").replace("|", " ")
    s = re.sub(r"[_\-–—]+", " ", s)          # -류 정리
    s = re.sub(r"[^\w가-힣\s]", " ", s)      # 특수문자 대충 제거(영문/숫자/한글/공백만)
    s = re.sub(r"\s+", " ", s).strip()

    # 4) 브랜드명이 제품명 앞에 있으면 제거 (이번엔 brand를 안 쓰니 실행되지 않음)
    if brand and not pd.isna(brand):
        brand_clean = str(brand).strip()
        if brand_clean:
            brand_pattern = re.escape(brand_clean)
            pattern = f"^{brand_pattern}\\s+"
            s = re.sub(pattern, "", s, flags=re.IGNORECASE)
            s = re.sub(r"\s+", " ", s).strip()

    return s


# =========================
# 2) CSV에 적용 (product_name 정규화)
# =========================
products_path = "crawling_data/review_only_100_data/table1_essence_reviews_259.csv"
df = pd.read_csv(products_path)

# product_name -> canonical_name
df["canonical_name"] = df["product_name"].apply(lambda x: normalize_product_name(x))

print("=" * 80)
print("product_name 정규화 결과 (상위 20개):")
print("=" * 80)

show_cols = [c for c in ["product_id", "product_name", "canonical_name"] if c in df.columns]
print(df[show_cols].head(20).to_string(index=False))

# 저장
out_path = "crawling_data/review_only_100_data/table1_essence_reviews_259.csv"
df.to_csv(out_path, index=False, encoding="utf-8-sig")
print(f"\n저장 완료: {out_path}")
```

```python
# user_keywords에서 특정 피부타입 키워드만 남기고 나머지 삭제

import pandas as pd

# 데이터 로드
reviews = pd.read_csv("crawling_data/review_only_100_data/table1_essence_reviews_259.csv")

print(f"원본 데이터 행 수: {len(reviews)}")
print(f"원본 user_keywords 샘플:")
print(reviews['user_keywords'].head(10))

# 유지할 키워드 목록
valid_keywords = ['지성', '건성', '복합성', '민감성', '약건성', '트러블성', '중성']

# user_keywords 전처리 함수
def filter_keywords(keywords_str):
    if pd.isna(keywords_str):
        return ''
    
    # '|'로 분리
    keywords = [k.strip() for k in str(keywords_str).split('|')]
    
    # 유효한 키워드만 필터링
    filtered = [k for k in keywords if k in valid_keywords]
    
    # 다시 '|'로 결합
    return ' | '.join(filtered) if filtered else ''

# user_keywords 필터링 적용
reviews['user_keywords'] = reviews['user_keywords'].apply(filter_keywords)

# 빈 문자열을 NaN으로 변환 (선택사항)
reviews['user_keywords'] = reviews['user_keywords'].replace('', pd.NA)

print(f"\n필터링 후 user_keywords 샘플:")
print(reviews['user_keywords'].head(10))

# 통계 확인
print(f"\n=== 통계 ===")
print(f"user_keywords가 비어있는 행: {reviews['user_keywords'].isna().sum()}개")
print(f"user_keywords가 있는 행: {reviews['user_keywords'].notna().sum()}개")

# 각 키워드별 개수
print(f"\n각 키워드 출현 횟수:")
for keyword in valid_keywords:
    count = reviews['user_keywords'].str.contains(keyword, na=False).sum()
    print(f"  {keyword}: {count}개")

# CSV 저장
reviews.to_csv('crawling_data/review_only_100_data/table1_essence_reviews_259.csv', index=False, encoding='utf-8-sig')

print("\n파일 저장 완료!")
```

```python
# user_keywords가 비어있는 상품을 reviews와 ingredients에서 제거하고 product_id 재할당

import pandas as pd

# 데이터 로드
reviews = pd.read_csv("crawling_data/review_only_100_data/table1_essence_reviews_259.csv")
ingredients = pd.read_csv("crawling_data/review_only_100_data/table3_essence_ingredients_259.csv")

print(f"원본 reviews 행 수: {len(reviews)}")
print(f"원본 ingredients 행 수: {len(ingredients)}")

# user_keywords가 비어있는 product_id 찾기
empty_keywords = reviews[reviews['user_keywords'].isna() | (reviews['user_keywords'] == '')]
products_to_remove = empty_keywords['product_id'].unique()

print(f"\n제거할 product_id: {products_to_remove}")
print(f"제거할 상품 수: {len(products_to_remove)}")

# reviews에서 해당 상품 제거
reviews_cleaned = reviews[~reviews['product_id'].isin(products_to_remove)].copy()

# ingredients에서 해당 상품 제거
ingredients_cleaned = ingredients[~ingredients['product_id'].isin(products_to_remove)].copy()

print(f"\n제거 후 reviews 행 수: {len(reviews_cleaned)}")
print(f"제거 후 ingredients 행 수: {len(ingredients_cleaned)}")

# ingredients를 product_id로 정렬 후 새로운 product_id 할당
ingredients_cleaned = ingredients_cleaned.sort_values('product_id').reset_index(drop=True)

# 기존 product_id와 새 product_id 매핑 딕셔너리 생성
old_to_new_id = {}
for idx, old_id in enumerate(ingredients_cleaned['product_id']):
    new_id = f'E{str(idx+1).zfill(3)}'
    old_to_new_id[old_id] = new_id

print(f"\n새로운 product_id 범위: E001 ~ E{str(len(ingredients_cleaned)).zfill(3)}")

# ingredients의 product_id 업데이트
ingredients_cleaned['product_id'] = ingredients_cleaned['product_id'].map(old_to_new_id)

# reviews의 product_id 업데이트
reviews_cleaned['product_id'] = reviews_cleaned['product_id'].map(old_to_new_id)

# 정렬 및 인덱스 리셋
ingredients_cleaned = ingredients_cleaned.reset_index(drop=True)
reviews_cleaned = reviews_cleaned.sort_values('product_id').reset_index(drop=True)

print(f"\n=== 최종 결과 ===")
print(f"ingredients 상품 수: {len(ingredients_cleaned)}")
print(f"reviews 리뷰 수: {len(reviews_cleaned)}")
print(f"reviews 고유 상품 수: {reviews_cleaned['product_id'].nunique()}")

print("\ningredients 첫 10개:")
print(ingredients_cleaned.head(10))

print("\nreviews 첫 10개:")
print(reviews_cleaned.head(10))

# CSV 저장
ingredients_cleaned.to_csv('crawling_data/review_only_100_data/table3_essence_ingredients_258.csv', index=False, encoding='utf-8-sig')
reviews_cleaned.to_csv('crawling_data/review_only_100_data/table1_essence_reviews_258.csv', index=False, encoding='utf-8-sig')

print("\n파일 저장 완료!")
print("- table3_essence_ingredients_258.csv")
print("- table1_essence_reviews_258.csv")
```

```python
# ingredients 성분 빈도 분석

import pandas as pd
import re
from collections import Counter

# pandas 출력 옵션 설정
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_colwidth', None)

# 데이터 로드
ingredients = pd.read_csv("crawling_data/review_only_100_data/table3_essence_ingredients_258.csv")

print(f"총 상품 수: {len(ingredients)}")
print(f"\n원본 ingredients 샘플:")
print(ingredients['ingredients'].head(3))

# 모든 성분 추출
all_ingredients = []

for ing_str in ingredients['ingredients'].dropna():
    # 쉼표로 분리
    ing_list = [i.strip() for i in str(ing_str).split(',')]
    all_ingredients.extend(ing_list)

print(f"\n총 성분 등장 횟수 (중복 포함): {len(all_ingredients)}")

# 성분 빈도 계산
ingredient_counts = Counter(all_ingredients)

print(f"고유 성분 수: {len(ingredient_counts)}")

# DataFrame으로 변환
ingredient_df = pd.DataFrame(ingredient_counts.items(), columns=['ingredient', 'count'])
ingredient_df = ingredient_df.sort_values('count', ascending=False).reset_index(drop=True)

print("\n=== 상위 50개 성분 (빈도순) ===")
print(ingredient_df.head(50))

# 통계
print(f"\n=== 통계 ===")
print(f"최대 빈도: {ingredient_df['count'].max()}회")
print(f"최소 빈도: {ingredient_df['count'].min()}회")
print(f"평균 빈도: {ingredient_df['count'].mean():.2f}회")
print(f"중앙값: {ingredient_df['count'].median():.0f}회")

# 빈도별 분포
print(f"\n=== 빈도별 성분 분포 ===")
print(f"100회 이상 등장: {len(ingredient_df[ingredient_df['count'] >= 100])}개")
print(f"50~99회 등장: {len(ingredient_df[(ingredient_df['count'] >= 50) & (ingredient_df['count'] < 100)])}개")
print(f"20~49회 등장: {len(ingredient_df[(ingredient_df['count'] >= 20) & (ingredient_df['count'] < 50)])}개")
print(f"10~19회 등장: {len(ingredient_df[(ingredient_df['count'] >= 10) & (ingredient_df['count'] < 20)])}개")
print(f"5~9회 등장: {len(ingredient_df[(ingredient_df['count'] >= 5) & (ingredient_df['count'] < 10)])}개")
print(f"1~4회 등장: {len(ingredient_df[ingredient_df['count'] < 5])}개")

# 상위 성분만 추출 (예: 10회 이상)
print(f"\n=== 10회 이상 등장하는 성분 ===")
frequent_ingredients = ingredient_df[ingredient_df['count'] >= 10]
print(f"개수: {len(frequent_ingredients)}개")
print(frequent_ingredients)

# CSV로 저장
ingredient_df.to_csv('crawling_data/review_only_100_data/ingredient_frequency_analysis.csv', index=False, encoding='utf-8-sig')
frequent_ingredients.to_csv('crawling_data/review_only_100_data/frequent_ingredients.csv', index=False, encoding='utf-8-sig')

print("\n파일 저장 완료!")
print("- ingredient_frequency_analysis.csv (전체 성분 빈도)")
print("- frequent_ingredients.csv (10회 이상 등장 성분)")

# 시각화를 위한 기본 정보
print(f"\n=== 시각화 추천 ===")
print("상위 20개 성분:")
print(ingredient_df.head(20)[['ingredient', 'count']])
```

```python
# GNN을 위한 성분 필터링 전략

import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# 성분 빈도 데이터 로드
ingredient_df = pd.read_csv(
    'crawling_data/review_only_100_data/ingredient_frequency_analysis.csv',
    encoding='utf-8-sig',
    low_memory=False,
    dtype=str  # 일단 모든 컬럼을 문자열로
)

print(f"총 고유 성분 수: {len(ingredient_df)}")
print(f"최대 빈도: {ingredient_df['count'].max()}")
print(f"최소 빈도: {ingredient_df['count'].min()}")
print(f"평균 빈도: {ingredient_df['count'].mean():.2f}")
print(f"중앙값: {ingredient_df['count'].median():.0f}")

# 빈도 분포 시각화 준비
print("\n=== 빈도 분포 ===")
bins = [0, 5, 10, 20, 50, 100, 300]
labels = ['1-4회', '5-9회', '10-19회', '20-49회', '50-99회', '100회+']

for i in range(len(bins)-1):
    count = len(ingredient_df[(ingredient_df['count'] >= bins[i]) & (ingredient_df['count'] < bins[i+1])])
    print(f"{labels[i]}: {count}개")

print("\n=== GNN을 위한 필터링 옵션 ===")

# 옵션 1: 극단값 제거 (너무 희귀하거나 너무 흔한 성분)
print("\n[옵션 1] 극단값 제거")
print("- 너무 희귀한 성분 (1-4회): 정보가 부족하고 노이즈가 될 수 있음")
print("- 너무 흔한 성분 (200회+): 모든 제품에 있어 변별력이 없을 수 있음")

option1_min = 5
option1_max = 200
option1 = ingredient_df[(ingredient_df['count'] >= option1_min) & (ingredient_df['count'] <= option1_max)]
print(f"  → {option1_min}~{option1_max}회 등장: {len(option1)}개 성분")

# 옵션 2: 중간 빈도 성분만 (의미있는 패턴)
print("\n[옵션 2] 중간 빈도 성분")
print("- 적절한 빈도 (10-150회): 충분한 정보와 변별력")

option2_min = 10
option2_max = 150
option2 = ingredient_df[(ingredient_df['count'] >= option2_min) & (ingredient_df['count'] <= option2_max)]
print(f"  → {option2_min}~{option2_max}회 등장: {len(option2)}개 성분")

# 옵션 3: 상위 N개 성분
print("\n[옵션 3] 상위 N개 성분")
for n in [50, 100, 150, 200]:
    print(f"  → 상위 {n}개: {n}개 성분 (최소 {ingredient_df.iloc[n-1]['count']}회 등장)")

# 옵션 4: 백분위수 기반
print("\n[옵션 4] 백분위수 기반")
percentiles = [25, 50, 75, 90, 95]
for p in percentiles:
    threshold = ingredient_df['count'].quantile(p/100)
    count = len(ingredient_df[ingredient_df['count'] >= threshold])
    print(f"  → 상위 {100-p}% 성분: {count}개 (빈도 {threshold:.0f}회 이상)")

# 추천 전략
print("\n" + "="*70)
print("📊 추천 전략")
print("="*70)

print("\n✅ 추천 1: 10회 이상 + 상위 150개")
recommended = ingredient_df[ingredient_df['count'] >= 10].head(150)
print(f"   → {len(recommended)}개 성분")
print(f"   → 빈도 범위: {recommended['count'].min()}~{recommended['count'].max()}회")
print("   → 장점: 충분한 변별력 + 적절한 노드 수")

print("\n✅ 추천 2: 20회 이상")
recommended2 = ingredient_df[ingredient_df['count'] >= 20]
print(f"   → {len(recommended2)}개 성분")
print(f"   → 빈도 범위: {recommended2['count'].min()}~{recommended2['count'].max()}회")
print("   → 장점: 더 안정적인 패턴, 노이즈 감소")

print("\n✅ 추천 3: 5회 이상 200회 이하")
recommended3 = ingredient_df[(ingredient_df['count'] >= 5) & (ingredient_df['count'] <= 200)]
print(f"   → {len(recommended3)}개 성분")
print(f"   → 빈도 범위: {recommended3['count'].min()}~{recommended3['count'].max()}회")
print("   → 장점: 넓은 범위, 다양한 성분 포함")

# 실제 필터링된 성분 저장
print("\n" + "="*70)
print("💾 필터링된 성분 저장")
print("="*70)

# 추천 1 저장
recommended.to_csv('crawling_data/review_only_100_data/ingredients_for_gnn_v1.csv', index=False, encoding='utf-8-sig')
print("✓ ingredients_for_gnn_v1.csv 저장 완료 (10회 이상 + 상위 150개)")

# 추천 2 저장
recommended2.to_csv('crawling_data/review_only_100_data/ingredients_for_gnn_v2.csv', index=False, encoding='utf-8-sig')
print("✓ ingredients_for_gnn_v2.csv 저장 완료 (20회 이상)")

# 추천 3 저장
recommended3.to_csv('crawling_data/review_only_100_data/ingredients_for_gnn_v3.csv', index=False, encoding='utf-8-sig')
print("✓ ingredients_for_gnn_v3.csv 저장 완료 (5회 이상 200회 이하)")

print("\n상위 20개 성분 (참고):")
print(ingredient_df.head(20))
```
