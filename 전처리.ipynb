```python
# canonical_name이 중복된 행들의 product_name을 전부 다 보이게 출력

import pandas as pd

# pandas 출력 옵션 설정 - 모든 행과 열을 다 보이게
pd.set_option('display.max_rows', None)  # 모든 행 표시
pd.set_option('display.max_columns', None)  # 모든 열 표시
pd.set_option('display.max_colwidth', None)  # 열 내용 전체 표시
pd.set_option('display.width', None)  # 출력 너비 제한 없음

# canonical_name이 중복된 것들만 필터링
duplicates = df[df.duplicated(subset=['canonical_name'], keep=False)]

# canonical_name으로 정렬해서 보기 쉽게
duplicates_sorted = duplicates.sort_values('canonical_name')

# 결과 출력
print(f"중복된 canonical_name 개수: {duplicates_sorted['canonical_name'].nunique()}")
print(f"중복된 총 행 개수: {len(duplicates_sorted)}")
print("\n중복된 행들:")
print(duplicates_sorted)

# canonical_name별 그룹으로 product_name 비교
print("\n\n=== canonical_name별 product_name 비교 ===")
for name, group in duplicates_sorted.groupby('canonical_name'):
    print(f"\n[{name}] - {len(group)}개 행")
    print(f"Product IDs: {group['product_id'].tolist()}")
    print("Product Names:")
    for idx, row in group.iterrows():
        print(f"  - {row['product_id']}: {row['product_name']}")
    print("-" * 100)
```

```python
# canonical_name이 중복된 행들 중 첫 번째(위쪽) 행만 남기고 나머지 삭제

import pandas as pd

# 원본 데이터프레임 확인
print(f"원본 행 개수: {len(df)}")

# 중복 확인
duplicates = df[df.duplicated(subset=['canonical_name'], keep=False)]
print(f"중복된 행 개수: {len(duplicates)}")

# canonical_name이 중복된 경우, 첫 번째 행만 유지하고 나머지 삭제
df_cleaned = df.drop_duplicates(subset=['canonical_name'], keep='first')

print(f"중복 제거 후 행 개수: {len(df_cleaned)}")
print(f"삭제된 행 개수: {len(df) - len(df_cleaned)}")

# 결과 확인
print("\n중복 제거 후 데이터:")
print(df_cleaned.head())

# 원본 df를 업데이트하려면 (선택사항):
# df = df_cleaned
```

```python
# df_cleaned의 상품명과 완전히 일치하는 리뷰만 필터링

import pandas as pd

# df_cleaned 읽기
df_cleaned = pd.read_csv('crawling_data/review_only_100_data/table1_essence_products_cleaned.csv')

# 원본 review 데이터 읽기 (필터링 전)
review = pd.read_csv("crawling_data/review_only_100_data/table1_essence_reviews.csv")

print(f"df_cleaned 상품 수: {len(df_cleaned)}")
print(f"원본 review 리뷰 수: {len(review)}")

# df_cleaned의 상품명 목록
valid_product_names = set(df_cleaned['product_name'])
print(f"유효한 상품명 수: {len(valid_product_names)}")

# review에서 상품명이 df_cleaned에 있는 것만 필터링
review_filtered = review[review['product_name'].isin(valid_product_names)].copy()

print(f"\n필터링 후 리뷰 수: {len(review_filtered)}")
print(f"제거된 리뷰 수: {len(review) - len(review_filtered)}")

# product_name을 기준으로 새로운 product_id 매핑
product_name_to_id = dict(zip(df_cleaned['product_name'], df_cleaned['product_id']))

# product_id 업데이트
review_filtered['product_id'] = review_filtered['product_name'].map(product_name_to_id)

# 매핑 확인
unmapped = review_filtered[review_filtered['product_id'].isna()]
if len(unmapped) > 0:
    print(f"\n경고: 매핑되지 않은 리뷰 {len(unmapped)}개")
else:
    print("\n모든 리뷰 매핑 성공!")

# product_id로 정렬 후 인덱스 리셋
review_filtered = review_filtered.sort_values('product_id').reset_index(drop=True)

# 결과 확인
print(f"\nreview_filtered에 있는 고유 상품 수: {review_filtered['product_id'].nunique()}")
print(f"첫 번째 product_id: {review_filtered['product_id'].iloc[0]}")
print(f"마지막 product_id: {review_filtered['product_id'].iloc[-1]}")

# 각 product_id별 리뷰 개수 확인
print("\nproduct_id별 리뷰 개수:")
print(review_filtered['product_id'].value_counts().sort_index())

# CSV 저장
review_filtered.to_csv('crawling_data/review_only_100_data/table1_essence_reviews_filtered.csv', index=False, encoding='utf-8-sig')

print("\n파일 저장 완료!")
```

```python
# review 데이터의 product_name을 table3_essence_ingredients.csv의 product_name으로 대체

import pandas as pd

# 데이터 로드
review = pd.read_csv("crawling_data/review_only_100_data/table1_essence_reviews.csv")
table3 = pd.read_csv("crawling_data/review_only_100_data/table3_esssence_ingredients.csv")

print(f"원본 review 리뷰 수: {len(review)}")
print(f"table3 상품 수: {len(table3)}")

# table3에서 product_id와 product_name 매핑 딕셔너리 생성
correct_product_names = dict(zip(table3['product_id'], table3['product_name']))

print(f"\n매핑 딕셔너리 생성: {len(correct_product_names)}개")

# review의 product_name을 올바른 이름으로 대체
review['product_name'] = review['product_id'].map(correct_product_names)

# 매핑되지 않은 항목 확인
unmapped = review[review['product_name'].isna()]
if len(unmapped) > 0:
    print(f"\n경고: 매핑되지 않은 리뷰 {len(unmapped)}개")
    print(f"매핑되지 않은 product_id: {unmapped['product_id'].unique()}")
else:
    print("\n모든 리뷰의 product_name 대체 성공!")

# 결과 확인
print(f"\nreview에 있는 고유 product_id 수: {review['product_id'].nunique()}")
print("\n수정된 review 데이터:")
print(review.head(10))

# CSV 저장
review.to_csv('crawling_data/review_only_100_data/table1_essence_reviews_corrected.csv', index=False, encoding='utf-8-sig')

print("\n파일 저장 완료: table1_essence_reviews_corrected.csv")
```
