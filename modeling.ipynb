# GNN

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F

# ====== 0) interactions ======
# final_total_review 컬럼: user_id, product_id (필수), user_rating (있으면 사용 가능 but 여기선 implicit)
inter = final_total_review[["user_id", "product_id"]].dropna().copy()

# 유저-상품 중복 리뷰가 여러 개면 1개로 압축(implicit이니까)
inter = inter.drop_duplicates(["user_id", "product_id"]).reset_index(drop=True)

print("interactions:", len(inter), "users:", inter["user_id"].nunique(), "items:", inter["product_id"].nunique())
```

```python
# ====== 1) id -> index ======
user2idx = {u:i for i,u in enumerate(inter["user_id"].unique())}
item2idx = {p:i for i,p in enumerate(inter["product_id"].unique())}

inter["u"] = inter["user_id"].map(user2idx).astype(int)
inter["i"] = inter["product_id"].map(item2idx).astype(int)

n_users = len(user2idx)
n_items = len(item2idx)
print(n_users, n_items)

# ====== (옵션) cold user/item 약간 필터(너무 sparse면 성능/평가 흔들림) ======
min_u = 2
min_i = 2

u_cnt = inter["u"].value_counts()
i_cnt = inter["i"].value_counts()

keep_u = set(u_cnt[u_cnt >= min_u].index)
keep_i = set(i_cnt[i_cnt >= min_i].index)

inter = inter[inter["u"].isin(keep_u) & inter["i"].isin(keep_i)].copy().reset_index(drop=True)

# 다시 인덱스 재매핑
user_ids = inter["u"].unique()
item_ids = inter["i"].unique()
u_remap = {old:new for new, old in enumerate(user_ids)}
i_remap = {old:new for new, old in enumerate(item_ids)}
inter["u"] = inter["u"].map(u_remap).astype(int)
inter["i"] = inter["i"].map(i_remap).astype(int)

n_users = inter["u"].nunique()
n_items = inter["i"].nunique()
print("after filter:", len(inter), n_users, n_items)
```

```python
from collections import defaultdict

rng = np.random.default_rng(42)

user_pos = defaultdict(list)
for u, i in zip(inter["u"].values, inter["i"].values):
    user_pos[u].append(i)

train_pos = defaultdict(set)
val_pos = defaultdict(set)
test_pos = defaultdict(set)

for u, items in user_pos.items():
    items = np.array(list(set(items)), dtype=int)
    rng.shuffle(items)
    if len(items) < 3:
        # 너무 적으면 train에만 (평가 불안정 방지)
        train_pos[u] = set(items.tolist())
        continue
    test_i = items[-1]
    val_i  = items[-2]
    tr_is  = items[:-2]
    train_pos[u] = set(tr_is.tolist())
    val_pos[u]   = {int(val_i)}
    test_pos[u]  = {int(test_i)}

# train edge 리스트 만들기
train_edges_u = []
train_edges_i = []
for u, items in train_pos.items():
    for i in items:
        train_edges_u.append(u)
        train_edges_i.append(i)

train_edges_u = np.array(train_edges_u, dtype=np.int64)
train_edges_i = np.array(train_edges_i, dtype=np.int64)

print("train edges:", len(train_edges_u))
print("val users:", sum(len(v)>0 for v in val_pos.values()), "test users:", sum(len(v)>0 for v in test_pos.values()))
```

```python
def build_norm_adj(n_users, n_items, users_np, items_np, device="cpu"):
    """
    Bipartite graph adjacency (user<->item) normalized: D^-0.5 A D^-0.5
    노드 수 = n_users + n_items
    """
    n_nodes = n_users + n_items

    u = torch.from_numpy(users_np).to(torch.long)
    i = torch.from_numpy(items_np).to(torch.long) + n_users  # item node offset

    # undirected edges: u->i, i->u
    row = torch.cat([u, i], dim=0)
    col = torch.cat([i, u], dim=0)
    val = torch.ones(row.size(0), dtype=torch.float32)

    # degree
    deg = torch.zeros(n_nodes, dtype=torch.float32)
    deg.scatter_add_(0, row, val)

    deg_inv_sqrt = torch.pow(deg, -0.5)
    deg_inv_sqrt[torch.isinf(deg_inv_sqrt)] = 0.0

    norm_val = deg_inv_sqrt[row] * val * deg_inv_sqrt[col]

    idx = torch.stack([row, col], dim=0)
    adj = torch.sparse_coo_tensor(idx, norm_val, (n_nodes, n_nodes)).coalesce().to(device)
    return adj

device = "cuda" if torch.cuda.is_available() else "cpu"
adj = build_norm_adj(n_users, n_items, train_edges_u, train_edges_i, device=device)
print(adj.shape, adj._nnz())
```

## LightGCN 모델 + BPR 학습

```python
class LightGCN(nn.Module):
    def __init__(self, n_users, n_items, emb_dim=64, n_layers=3):
        super().__init__()
        self.n_users = n_users
        self.n_items = n_items
        self.n_layers = n_layers
        self.user_emb = nn.Embedding(n_users, emb_dim)
        self.item_emb = nn.Embedding(n_items, emb_dim)
        nn.init.normal_(self.user_emb.weight, std=0.01)
        nn.init.normal_(self.item_emb.weight, std=0.01)

    def forward(self, adj):
        # initial embeddings
        x0 = torch.cat([self.user_emb.weight, self.item_emb.weight], dim=0)  # (n_nodes, d)
        xs = [x0]
        x = x0
        for _ in range(self.n_layers):
            x = torch.sparse.mm(adj, x)
            xs.append(x)
        x_final = torch.mean(torch.stack(xs, dim=0), dim=0)
        user_final = x_final[:self.n_users]
        item_final = x_final[self.n_users:]
        return user_final, item_final

def bpr_loss(u_emb, pos_emb, neg_emb):
    pos_scores = (u_emb * pos_emb).sum(dim=1)
    neg_scores = (u_emb * neg_emb).sum(dim=1)
    return -torch.log(torch.sigmoid(pos_scores - neg_scores) + 1e-8).mean()

# negative sampling helper
train_pos_sets = {u:set(items) for u, items in train_pos.items()}

def sample_batch(batch_size=1024):
    users = np.random.choice(list(train_pos_sets.keys()), size=batch_size, replace=True)
    pos = []
    neg = []
    for u in users:
        pos_i = np.random.choice(list(train_pos_sets[u]))
        # neg: 랜덤으로 뽑되 pos에 없을 때까지
        while True:
            neg_i = np.random.randint(0, n_items)
            if neg_i not in train_pos_sets[u]:
                break
        pos.append(pos_i)
        neg.append(neg_i)
    return (torch.tensor(users, dtype=torch.long, device=device),
            torch.tensor(pos, dtype=torch.long, device=device),
            torch.tensor(neg, dtype=torch.long, device=device))

model = LightGCN(n_users, n_items, emb_dim=64, n_layers=3).to(device)
opt = torch.optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)

# train
epochs = 30
steps_per_epoch = 200  # 데이터 크기에 맞게 조절
for ep in range(1, epochs+1):
    model.train()
    total = 0.0
    for _ in range(steps_per_epoch):
        u, pi, ni = sample_batch(batch_size=1024)
        user_f, item_f = model(adj)
        loss = bpr_loss(user_f[u], item_f[pi], item_f[ni])
        opt.zero_grad()
        loss.backward()
        opt.step()
        total += loss.item()
    print(f"epoch {ep:02d} | loss {total/steps_per_epoch:.4f}")
```

## 평가 - Recall@K / NDCG@K (test 기준)

```python
def recall_ndcg_at_k(model, adj, train_pos, test_pos, k=20):
    model.eval()
    with torch.no_grad():
        user_f, item_f = model(adj)  # (U,d), (I,d)
        user_f = user_f.cpu()
        item_f = item_f.cpu()

    recalls = []
    ndcgs = []

    all_items = np.arange(item_f.size(0))

    for u, gt_items in test_pos.items():
        if len(gt_items) == 0:
            continue
        gt = list(gt_items)

        scores = (user_f[u].unsqueeze(0) @ item_f.T).squeeze(0).numpy()

        # mask train items (이미 본 건 추천에서 제외)
        for seen in train_pos.get(u, set()):
            scores[seen] = -1e9

        topk = np.argpartition(-scores, k)[:k]
        topk = topk[np.argsort(-scores[topk])]

        hit = 0
        dcg = 0.0
        for rank, item in enumerate(topk, start=1):
            if item in gt:
                hit += 1
                dcg += 1.0 / np.log2(rank + 1)

        recall = hit / len(gt)
        idcg = sum(1.0 / np.log2(r + 1) for r in range(1, min(len(gt), k) + 1))
        ndcg = (dcg / idcg) if idcg > 0 else 0.0

        recalls.append(recall)
        ndcgs.append(ndcg)

    return float(np.mean(recalls)) if recalls else 0.0, float(np.mean(ndcgs)) if ndcgs else 0.0

r20, n20 = recall_ndcg_at_k(model, adj, train_pos, test_pos, k=20)
print("Recall@20:", r20, "NDCG@20:", n20)
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD

# item2idx / idx2item / n_items 이미 만들어둔 상태라고 가정
idx2item = {v:k for k,v in item2idx.items()}

# product_id -> ingredients(list) 매핑
prod_ing_map = final_products.set_index("product_id")["ingredients"].to_dict()

corpus = []
missing = 0

for item_idx in range(n_items):
    pid = idx2item[item_idx]
    ings = prod_ing_map.get(pid, [])
    if not isinstance(ings, list):
        missing += 1
        ings = []
    # 리스트 -> 문서 문자열
    corpus.append(" ".join([str(x).strip() for x in ings if str(x).strip()]))

print("missing items:", missing, "/", n_items)

tfidf = TfidfVectorizer(token_pattern=r"(?u)\b\w+\b", min_df=2)
X = tfidf.fit_transform(corpus)

feat_dim = 64
svd = TruncatedSVD(n_components=feat_dim, random_state=42)
X_dense = svd.fit_transform(X).astype(np.float32)

# 정규화(추천)
X_dense = X_dense / (np.linalg.norm(X_dense, axis=1, keepdims=True) + 1e-12)

print("X_dense:", X_dense.shape)  # (n_items, feat_dim)
```
