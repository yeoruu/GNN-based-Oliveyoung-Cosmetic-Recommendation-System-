```python
import pandas as pd

df = pd.read_csv("/Users/jeong-yeon-u/Desktop/GNN-based-Oliveyoung-Cosmetic-Recommendation-System-/products_essence_final2.csv")

cols = [
    "product_id",
    "category",
    "brand",
    "canonical_name",
    "product_rating",
    "ingredients_normalized_list"
]

df_selected = df[cols]
df_selected = df_selected.rename(columns={
    "canonical_name": "product_name",
    "ingredients_normalized_list": "ingredients"
})
df_selected.to_csv("products_essence_final_only_cols.csv", index=False, encoding="utf-8-sig")
print("저장완료")
```
